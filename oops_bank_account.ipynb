```python
import random

class BankAccount:
    def __init__(self, name, balance, pin):
        self.account_number = str(random.randint(10**6, 10**7 - 1))
        self.name = name
        self.balance = balance
        self.pin = pin

    def deposit(self, amt):
        self.balance += amt
        print("Deposited:", amt)

    def withdraw(self, amt):
        if input("Enter PIN: ") == self.pin:
            if amt <= self.balance:
                self.balance -= amt
                print("Withdrawn:", amt)
            else:
                print("Insufficient balance")
        else:
            print("Wrong PIN")

class SavingAccount(BankAccount):
    rate = 4
    def apply_interest(self):
        interest = self.balance * self.rate / 100
        self.balance += interest
        print("Interest added:", interest)

class CurrentAccount(BankAccount):
    limit = 50000
    def withdraw(self, amt):
        if input("Enter PIN: ") == self.pin:
            if amt <= self.balance + self.limit:
                self.balance -= amt
                print("Withdrawn:", amt)
            else:
                print("Overdraft limit exceeded")
        else:
            print("Wrong PIN")

acc_type = input("Enter account type (saving/current): ").lower()
name = input("Enter account holder name: ")
balance = float(input("Enter initial balance: "))
pin = input("Set a 4-digit PIN: ")

if acc_type == "saving":
    acc = SavingAccount(name, balance, pin)
    print("Account Number:", acc.account_number)
    print("Interest Rate:", acc.rate, "%")
elif acc_type == "current":
    acc = CurrentAccount(name, balance, pin)
    print("Account Number:", acc.account_number)
    print("Overdraft Limit: â‚¹", acc.limit)
else:
    print("Invalid account type")
    exit()

while True:
    if acc_type == "saving":
        print("\n1.Deposit\n  2.Withdraw\n  3.Balance\n  4.Apply Interest\n  5.Exit")
    else:
        print("\n1.Deposit\n  2.Withdraw\n  3.Balance\n  5.Exit")
    ch = input("Enter choice: ")
    if ch == "1":
        amt = float(input("Amount: "))
        acc.deposit(amt)
    elif ch == "2":
        amt = float(input("Amount: "))
        acc.withdraw(amt)
    elif ch == "3":
        print("Balance:", acc.balance)
    elif ch == "4" and acc_type == "saving":
        acc.apply_interest()
    elif ch == "5":
        break
    else:
        print("Invalid choice")



```
